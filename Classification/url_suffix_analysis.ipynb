```python
import pandas as pd
import os
from urllib.parse import urlparse
from collections import Counter

pd.set_option("display.max_colwidth", 200)
```

```python
# Path dataset (chỉnh lại nếu cần)
DATASET_PATH = "D:\\0_Daniel\\HK5\\IE105\\3_DoAn\\dataset.csv"

df = pd.read_csv(DATASET_PATH)

# Chỉ cần cột url
df = df[['url']].dropna().reset_index(drop=True)

df.head()
```

```python
def extract_suffix_parts(url):
    parsed = urlparse(str(url).lower())
    path = parsed.path
    filename = os.path.basename(path)
    extension = os.path.splitext(filename)[1]
    return path, filename, extension

df[['path', 'filename', 'extension']] = df['url'].apply(
    lambda x: pd.Series(extract_suffix_parts(x))
)

df.head(10)
```

```python
ext_counter = Counter(df['extension'])

ext_df = pd.DataFrame(
    ext_counter.most_common(),
    columns=['extension', 'count']
)

ext_df.head(20)
```

```python
filename_counter = Counter(df['filename'])

filename_df = pd.DataFrame(
    filename_counter.most_common(),
    columns=['filename', 'count']
)

filename_df.head(20)
```

```python
from collections import Counter
import os
from urllib.parse import urlparse

# Tách filename từ URL
df['filename'] = df['url'].apply(
    lambda x: os.path.basename(urlparse(str(x)).path)
)

# Đếm số lần xuất hiện
filename_counter = Counter(df['filename'])

# Chuyển sang DataFrame cho dễ nhìn
filename_df = (
    pd.DataFrame(filename_counter.items(), columns=['filename', 'count'])
    .sort_values(by='count', ascending=False)
    .reset_index(drop=True)
)

filename_df
```

```python
from collections import Counter

filename_counter = Counter(df['filename'])

filename_3plus = (
    pd.DataFrame(filename_counter.items(), columns=['filename', 'count'])
    .query("count >= 3")
    .sort_values(by='count', ascending=False)
    .reset_index(drop=True)
)

filename_3plus

filename_3plus.to_csv("filename_appear_3plus.csv", index=False)
```
